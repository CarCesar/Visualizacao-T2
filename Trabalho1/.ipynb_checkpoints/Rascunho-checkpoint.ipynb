```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
tudo= pd.read_csv('Dados\dados.csv',sep=';').rename(columns = {'ano':'year','IDH':'HDI'})
tudo
```

```python
select_year = alt.selection_single(
    name='Select', fields=['year'], init={'year': 2019},
    bind=alt.binding_range(min=2011, max=2019, step=2)
)
selection = alt.selection_multi(fields=['continent'], bind='legend')
scales = alt.selection_interval(bind='scales')

base_int = alt.Chart(tudo, title ='Corruption and Human Development').encode(
    x = alt.X('CPI:Q',title='Corruption Percent Index (10 = least corrupt)',
              scale=alt.Scale(zero=False),axis=alt.Axis(grid=False)),
    y = alt.Y('HDI:Q', title = 'Human Development Index (1=best)',scale=alt.Scale(zero=False,domain =[.3,1])),
    ).properties(
    width=700,
    height=300,
)

points_int = base_int.mark_point(size=70).encode(color = alt.Color('continent',scale=alt.Scale(
        range =[ "#2CD602","#D625C4",'#0F9FEE',"#F5B938", "#FF1C14", "#A85402", "#ff6347"]
        ),legend=alt.Legend(
        orient='none',
        legendY=-20,
        legendX=193,
        direction='horizontal',
        title=None,
        labelFontSize = 11)),
        opacity=alt.condition(selection, alt.value(.9), alt.value(0.1)),                                                       
        tooltip =['country','CPI','HDI','continent','year']
        ).add_selection(
    selection,
    scales,
    select_year
    ).transform_filter(select_year)

curve_int = base_int.mark_line(
    color='red',
    size=3,
    opacity=0.25
).transform_regression(
    'CPI',    
    'HDI',
    method='log'
).encode()

curve_int + points_int 
```

```python
america = alt.Chart(tudo[tudo['continent']=='Americas']
    ).encode(
        x = alt.X('CPI:Q',
                  title='Corruption Percent Index (10 = least corrupt)',
                  scale=alt.Scale(zero=False),
                  axis=alt.Axis(grid=False)
                 ),
        y = alt.Y('HDI:Q',
                  title = 'Human Development Index (1=best)',
                  scale=alt.Scale(zero=False,
                                  domain =[.3,1]
                                 )
                 ),
    ).properties(
        width = 700,
        height = 300
)

americaP = america.mark_point().encode(color = alt.Color('sub_region'),opacity='year')




americaP
```

```python
input_cont = alt.binding_select(options=['Africa','Americas','Asia','Europe','Oceania'])
select_cont = alt.selection_single(fields=['continent'], bind=input_cont, name='Regions of')
select_reg = alt.selection_multi(fields=['sub_region'], bind='legend')

alt.Chart(tudo).mark_point(size = 70).encode(
    x = alt.X('CPI:Q',title='Corruption Percent Index (10 = least corrupt)',
              scale=alt.Scale(zero=False),axis=alt.Axis(grid=False)),
    y = alt.Y('HDI:Q', title = 'Human Development Index (1=best)',scale=alt.Scale(zero=False,domain =[.3,1])),
    color=alt.Color('sub_region:N',legend=alt.Legend(title=None)),
    opacity = alt.condition(select_reg, alt.value(1),
                            alt.value(.1)),
    shape = alt.Shape('year:N'),
    fill = alt.Fill('country',legend = None),
    tooltip=['country','year','sub_region','CPI',"HDI"]
).add_selection(
    select_cont,
    select_reg
).transform_filter(
    select_cont,
).properties(
    width=700,
    height=300,
)
```

```python
lista= []
for c in tudo['continent'].unique():
    input_cont = alt.binding_select(options=tudo['sub_region'][tudo['continent']==c].unique())
    select_cont = alt.selection_single(fields=['sub_region'], bind=input_cont, 
                                       name='Regions of',
                                       init={'sub_region':tudo['sub_region'][tudo['continent']==c].unique()[0]})
    select_reg = alt.selection_multi(fields=['country'], bind='legend')

    cont = alt.Chart(tudo[tudo['continent'] == c],title=c).encode(
        x = alt.X('CPI:Q',title='Corruption Percent Index (10 = least corrupt)',
                  scale=alt.Scale(zero=False,domain =[0,10]),axis=alt.Axis(grid=False)),
        y = alt.Y('HDI:Q', title = 'Human Development Index (1=best)',scale=alt.Scale(zero=False,domain =[.3,1])),
    ).properties(
        width=700,
        height=300,
    )

    graph = cont.mark_point(size = 70).encode(color=alt.Color('country:N',legend=alt.Legend(
            orient='none',
            legendY=-20,
            legendX=-40,
            direction='horizontal',
            title=None,
            labelFontSize = 7)),
        opacity = alt.condition(select_reg, alt.value(1),
                                alt.value(0.1)),
        size = alt.Opacity('year:N',scale = alt.Scale(domain=[2011,2013,2015,2019],
                                                     range=[11,33,55,77]),
                                                     legend = alt.Legend(
                                                                         orient = 'bottom-right',
                                                                         title = None)
                          ),
        #shape = alt.Shape('ano:N'),
        fill = 'country',
        #fillOpacity = 'ano:T',
        tooltip=['country','year','sub_region','CPI',"HDI"]
    ).add_selection(
        select_cont,
        select_reg
    ).transform_filter(
        select_cont,
    ) 
    lista.append(graph)
```

```python
input_contEuro = alt.binding_select(options=tudo['sub_region'][tudo['continent']=='Europe'].unique())
select_contEuro = alt.selection_single(fields=['sub_region'], bind=input_contEuro, 
                                   name='Regions of',
                                   init={'sub_region':tudo['sub_region'][tudo['continent']=="Europe"].unique()[0]})
select_regEuro = alt.selection_multi(fields=['country'], bind='legend')

contEuro = alt.Chart(tudo[tudo['continent'] == "Europe"],title="Europe").encode(
    x = alt.X('CPI:Q',title='Corruption Percent Index (10 = least corrupt)',
              scale=alt.Scale(zero=False,domain =[0,10]),axis=alt.Axis(grid=False)),
    y = alt.Y('HDI:Q', title = 'Human Development Index (1=best)',scale=alt.Scale(zero=False,domain =[.3,1])),
).properties(
    width=700,
    height=300,
)

graphEuro = contEuro.mark_point(size = 70).encode(color=alt.Color('country:N',legend=alt.Legend(
        orient='none',
        legendY=-20,
        legendX=-40,
        direction='horizontal',
        title=None,
        labelFontSize = 7)),
    opacity = alt.condition(select_regEuro, alt.value(1),
                            alt.value(0.1)),
    size = alt.Opacity('year:N',scale = alt.Scale(domain=[2011,2013,2015,2019],
                                                 range=[11,40,65,99]),
                                                 legend = alt.Legend(
                                                                     orient = 'bottom-right',
                                                                     title = None)
                      ),
    fill = 'country',
    tooltip=['country','year','sub_region','CPI',"HDI"]
).add_selection(
    select_contEuro,
    select_regEuro
).transform_filter(
    select_contEuro,
)

linhaEURO = contEuro.mark_line(
    color='red',
    size=3,
    opacity=0.25
).transform_regression(
    'CPI',    
    'HDI',
    method='log'
).encode(
opacity=alt.Opacity('continent',legend=alt.Legend(orient='none',legendY=7,legendX=17,direction='horizontal',
        title=None,values=['Europa']),scale=alt.Scale(type='ordinal', range=[.25])))

linhageral = alt.Chart(tudo).mark_line(color = 'grey').encode(
    x = alt.X('CPI:Q',title='Corruption Percent Index (10 = least corrupt)',
              scale=alt.Scale(zero=False,domain =[0,10]),axis=alt.Axis(grid=False)),
    y = alt.Y('HDI:Q', title = 'Human Development Index (1=best)',scale=alt.Scale(zero=False,domain =[.3,1])),
    opacity=alt.Opacity('line',legend=alt.Legend(orient='top-left',legendY=20,legendX=50,direction='horizontal',
        title=None,values=['World']),scale=alt.Scale(type='ordinal', range=[0.25]))
).transform_regression(
    'CPI',    
    'HDI',
    method='log'
).properties(
    width=700,
    height=300,
)

linhaEURO + graphEuro + linhageral
```

```python
for c in tudo['continent'].unique():
    input_contEuro = alt.binding_select(options=tudo['sub_region'][tudo['continent']==c].unique())
    select_contEuro = alt.selection_single(fields=['sub_region'], bind=input_contEuro, 
                                       name='Regions of',
                                       init={'sub_region':tudo['sub_region'][tudo['continent']==c].unique()[0]})
    select_regEuro = alt.selection_multi(fields=['country'], bind='legend')

    contEuro = alt.Chart(tudo[tudo['continent'] == c],title=c).encode(
        x = alt.X('CPI:Q',title='Corruption Percent Index (10 = least corrupt)',
                  scale=alt.Scale(zero=False,domain =[0,10]),axis=alt.Axis(grid=False)),
        y = alt.Y('HDI:Q', title = 'Human Development Index (1=best)',scale=alt.Scale(zero=False,domain =[.3,1])),
    ).properties(
        width=700,
        height=300,
    )

    graphEuro = contEuro.mark_point(size = 70).encode(color=alt.Color('country:N',legend=alt.Legend(
            orient='none',
            legendY=-20,
            legendX=-40,
            direction='horizontal',
            title=None,
            labelFontSize = 7)),
        opacity = alt.condition(select_regEuro, alt.value(1),
                                alt.value(0.1)),
        size = alt.Opacity('year:N',scale = alt.Scale(domain=[2011,2013,2015,2019],
                                                     range=[11,40,65,99]),
                                                     legend = alt.Legend(
                                                                         orient = 'bottom-right',
                                                                         title = None)
                          ),
        fill = 'country',
        tooltip=['country','year','sub_region','CPI',"HDI"]
    ).add_selection(
        select_contEuro,
        select_regEuro
    ).transform_filter(
        select_contEuro,
    )

    linhaEURO = contEuro.mark_line(
        color='red',
        size=3,
        opacity=0.25
    ).transform_regression(
        'CPI',    
        'HDI',
        method='log'
    ).encode(
    opacity=alt.Opacity('continent',legend=alt.Legend(orient='none',legendY=7,legendX=17,direction='horizontal',
            title=None,values=[c]),scale=alt.Scale(type='ordinal', range=[.25])))

    linhageral = alt.Chart(tudo).mark_line(color = 'grey').encode(
        x = alt.X('CPI:Q',title='Corruption Percent Index (10 = least corrupt)',
                  scale=alt.Scale(zero=False,domain =[0,10]),axis=alt.Axis(grid=False)),
        y = alt.Y('HDI:Q', title = 'Human Development Index (1=best)',scale=alt.Scale(zero=False,domain =[.3,1])),
        opacity=alt.Opacity('line',legend=alt.Legend(orient='top-left',legendY=20,legendX=50,direction='horizontal',
            title=None,values=['World']),scale=alt.Scale(type='ordinal', range=[0.25]))
    ).transform_regression(
        'CPI',    
        'HDI',
        method='log'
    ).properties(
        width=700,
        height=300,
    )

    graficopronto = linhaEURO + graphEuro + linhageral
    nome = c + '.html'
    graficopronto.save(nome)
    #Coloquei eles na pasta HTML para limpar o diretorio
```
