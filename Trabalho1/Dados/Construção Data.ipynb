```python
import pandas as pd
import numpy as np
import re
```

```python
a = open('idh.csv','r')
conteudo = a.readlines()
a.close()
```

```python
c = conteudo
lista = []
for b in c:
    b = re.sub('"','',b)
    d = b.split(',')
    e = []
    for f in d:
        e.append(f.strip())
    lista.append(e)
```

```python
idh = pd.DataFrame(lista[1:]).drop(62,axis=1)
```

```python
idh.columns = lista[0]
```

```python
gini = pd.read_csv('Gini2011.csv')
```

```python
for novo in mapa:
    idh.loc[idh['Country']==mapa[novo],'Country'] = novo
```

```python
df=pd.merge(gini[['country','region','score']],
            idh[['Country','2011']].rename(columns={'Country':'country'}),
            how='inner',on = 'country')
df
```

```python
sig = ["WE/EU", "AP", "AME", "MENA", "SSA", "ECA"]
nome = ["OECD", "Asia & Oceania", "Americas", 
                  "Middle East & North Africa", "Sub-Saharan Africa", 
                  "Eastern & Central Europe"]
for novo in range(0,len(sig)):
    df.loc[df['region']==sig[novo],'region'] = nome[novo]
```

```python
df['ano']=177*[2011]
```

```python
df = df.rename(columns={'score':'CPI','2011':'IDH'})
df.info()
```

```python
df['IDH'] = pd.to_numeric(df['IDH'])
df.info()
```

```python
print(df['IDH'][df['IDH'].isnull()])
df.drop(index=69,inplace=True)
# não tinha corrupção
```
